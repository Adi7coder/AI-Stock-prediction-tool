```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf

nse_tickers = ["RELIANCE.NS", "TCS.NS", "INFY.NS", "SBIN.NS", "HDFCBANK.NS", "ICICIBANK.NS", "HINDUNILVR.NS"]

# download
data = yf.download(nse_tickers, start="2015-01-01", end="2024-07-05")["Close"]
data = data.ffill()

# plot
plt.figure(figsize=(12,6))
for stock in data.columns:
    plt.plot(data.index, data[stock], label=stock)
plt.title("NSE Stock Closing Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price (INR)")
plt.legend()
plt.show()

# correlation
corr = data.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Among NSE Stocks")
plt.show()

# describe
print(data.describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

tickers = ["RELIANCE", "TCS", "INFY", "SBIN", "HDFCBANK", "ICICIBANK", "HINDUNILVR"]

for ticker in tickers:
    file_path = f"../data/{ticker}.csv"
    df = pd.read_csv(file_path, index_col=0, parse_dates=True)
    
    if df.empty or "Close" not in df.columns:
        print(f"{ticker}: file is empty or missing Close column, skipping.")
        continue

    print(f"\nPreview of {ticker}:\n", df.head())

    # convert to numeric if needed
    df["Close"] = pd.to_numeric(df["Close"], errors="coerce")
    df.dropna(subset=["Close"], inplace=True)

    if df.empty:
        print(f"{ticker}: no valid numeric Close data after cleaning, skipping.")
        continue
    
    # rolling SMA
    df["Close_SMA30"] = df["Close"].rolling(window=30).mean()

    plt.figure(figsize=(14, 6))
    plt.plot(df.index, df["Close"], color="green", linewidth=1.5, alpha=0.6, label="Close Price")
    plt.plot(df.index, df["Close_SMA30"], color="blue", linewidth=2, label="30-day SMA")

    plt.title(f"{ticker} Closing Price (with 30-day SMA)", fontsize=14)
    plt.ylabel("Price (INR)", fontsize=12)
    plt.xlabel("Date", fontsize=12)
    plt.legend()
    plt.grid(True, linestyle="--", alpha=0.5)
    plt.tight_layout()
    plt.show()
```
