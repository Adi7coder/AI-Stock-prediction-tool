```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf


tickers = ["AAPL", "MSFT", "TSLA","GOOGL","AMZN","NVDA","META"]

data = yf.download(tickers, start="2015-01-01", end="2025-07-01")["Close"]
data = data.ffill()


plt.figure(figsize=(12,6))
for stock in data.columns:
    plt.plot(data.index, data[stock], label=stock)
plt.title("Stock Closing Prices Over Time")
plt.xlabel("Date")
plt.ylabel("Close Price")
plt.legend()
plt.show()


corr = data.corr()
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.title("Correlation Among the Stocks that We have Chosen")
plt.show()


print(data.describe())
```

```python
for ticker in tickers:
    print(f"Fetching data for {ticker}...")
    df = yf.download(ticker, start="2015-01-01", end="2025-07-01")[["Close"]]


    df.columns = [f"{ticker}_Close"] 
    df.to_csv(f"../data/{ticker}.csv")
    print(f"Saved data/{ticker}.csv")
```

```python
for ticker in tickers:
    df = pd.read_csv(f"../data/{ticker}.csv", index_col=0, parse_dates=True)
    plt.figure(figsize=(10,4))
    plt.plot(df)
    plt.title(f"{ticker} Closing Price")
    plt.ylabel("Price")
    plt.xlabel("Date")
    plt.show()
```
